## 4. 연습

EPL 리그 테이블 만들기
- 순위, 팀명, 승, 무, 패, 승점, 득점, 실점, 득실차

```python
target_season = '2008/2009'
england_id=country_t[country_t['name']=='England']['id'].values[0]
eng_match=match_t[(match_t['country_id']==england_id) & (match_t['season']==target_season)]
eng_match=eng_match[['home_team_api_id', 'away_team_api_id', 'home_team_goal', 'away_team_goal']]

home_eng_team = team_t.copy()
home_eng_team.rename(columns = {'team_api_id': "home_team_api_id"}, inplace=True)
home_eng_team = home_eng_team[['home_team_api_id', 'team_long_name']]

away_eng_team = team_t.copy()
away_eng_team.rename(columns = {'team_api_id': "away_team_api_id"}, inplace=True)
away_eng_team = away_eng_team[['away_team_api_id', 'team_long_name']]

target = pd.merge(eng_match, home_eng_team, how="left", on='home_team_api_id')
target.rename(columns={'team_long_name':"home_team_name"}, inplace=True)
target = pd.merge(target, away_eng_team, how='left', on='away_team_api_id')
target.rename(columns={'team_long_name':"away_team_name"}, inplace=True)
target = target[["home_team_name", "away_team_name", "home_team_goal", "away_team_goal"]]
target.head()
```

```python
def team_point(row, home=True):
  
  my_goal = row['home_team_goal'] if home else row['away_team_goal']
  your_goal = row['away_team_goal'] if home else row['home_team_goal']

  if my_goal > your_goal:
    return 3
  elif my_goal == your_goal:
    return 1
  else:
    return 0

target['home_team_point'] = target.apply(lambda x: team_point(x, home=True), axis=1)
target['away_team_point'] = target.apply(lambda x: team_point(x, home=False), axis=1)

team_list = list(target['home_team_name'].unique())


data = []
for team in team_list:
  team_table = target[(target['home_team_name']==team) | (target['away_team_name']==team)]
  home_wins = len(team_table[(team_table['home_team_name']==team) & (team_table['home_team_point']==3)])
  home_draws = len(team_table[(team_table['home_team_name']==team) & (team_table['home_team_point']==1)])
  home_loses = len(team_table[(team_table['home_team_name']==team) & (team_table['home_team_point']==0)])
  home_scores = sum(team_table[(team_table['home_team_name']==team)]['home_team_goal'].values)
  home_againsts = sum(team_table[(team_table['home_team_name']==team)]['away_team_goal'].values)

  away_wins = len(team_table[(team_table['away_team_name']==team) & (team_table['away_team_point']==3)])
  away_draws = len(team_table[(team_table['away_team_name']==team) & (team_table['away_team_point']==1)])
  away_loses = len(team_table[(team_table['away_team_name']==team) & (team_table['away_team_point']==0)])
  away_scores = sum(team_table[(team_table['away_team_name']==team)]['away_team_goal'].values)
  away_againsts = sum(team_table[(team_table['away_team_name']==team)]['home_team_goal'].values)

  goal_for = home_scores+away_scores
  goal_against = home_againsts+away_againsts

  home_points = 3*home_wins + home_draws
  away_points = 3*away_wins + away_draws
  data.append([team, len(team_table), home_wins+away_wins, home_draws+away_draws, home_loses+away_loses, home_points+away_points, \
               goal_for, goal_against, goal_for-goal_against])

league_table = pd.DataFrame(data, columns=['팀', '경기', '승', '무', '패', '승점', '득점', '실점', "득실"])
league_table = league_table.sort_values(by=["승점", "득실", "득점"] , ascending=False).reset_index(drop=True)
league_table.index += 1
league_table
```
